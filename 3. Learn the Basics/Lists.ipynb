Introduction to Lists in Python
In this tutorial, we'll learn about lists in Python and how to use them in Google Colab. Lists are a fundamental data structure in Python that allows you to store ordered collections of items.

What is a List?
A list is an ordered collection of items that can be of any type. Lists are created by placing the items inside square brackets [], separated by commas.

For example:

```python
my_list = [1, 2, 3, "apple", "banana", True]
```

Here, my_list is a list containing three integers, two strings, and a boolean.

Using Lists in Google Colab
To use lists in Google Colab, you can define them in a code cell and manipulate them using various list methods and operations.

For example, create a new code cell in your Google Colab notebook and type the following code:

```python
# Creating a list
fruits = ["apple", "banana", "cherry"]

# Accessing list items
print(fruits[0])  # First item
print(fruits[-1]) # Last item

# Adding an item to the end of the list
fruits.append("orange")

# Inserting an item at a specific position
fruits.insert(1, "blueberry")

# Removing an item
fruits.remove("banana")

# Printing the modified list
print(fruits)
```

Looping Through a List
You can use a for loop to iterate over the items in a list:

```python
# Looping through a list
for fruit in fruits:
    print(fruit)
```

List Comprehensions
List comprehensions provide a concise way to create lists. They consist of square brackets containing an expression followed by a for clause:

```python
# Creating a new list with list comprehension
squares = [x**2 for x in range(10)]
print(squares)
```
