Introduction to If Statements in Python
In this tutorial, we'll learn about if statements in Python and how to use them in Google Colab. If statements are a fundamental concept in programming that allows you to control the flow of your code based on conditions.

What is an If Statement?
An if statement is used to test a condition and execute a block of code if the condition is true. In Python, the syntax for an if statement is as follows:

if condition:
    # code to execute if condition is true


You can also include elif (else if) and else blocks to handle different conditions:



if condition1:
    # code to execute if condition1 is true
elif condition2:
    # code to execute if condition2 is true
else:
    # code to execute if none of the above conditions are true


Using If Statements in Google Colab
To use if statements in Google Colab, you can write your Python code in a code cell and execute it.

For example, create a new code cell in your Google Colab notebook and type the following code:

```python
# Using if statements
age = 20

if age < 18:
    print("You are a minor.")
elif age >= 18 and age < 60:
    print("You are an adult.")
else:
    print("You are a senior.")
```

Nested If Statements
You can also nest if statements within each other to check multiple conditions:

```python
# Nested if statements
age = 25
is_student = True

if age < 30:
    if is_student:
        print("You are a young student.")
    else:
        print("You are a young adult.")
else:
    print("You are over 30 years old.")
```
