Introduction to Functions in Python
In this tutorial, we'll learn about functions in Python and how to use them in Google Colab. Functions are a fundamental concept in programming that allow you to organize your code into reusable blocks.

What is a Function?
A function is a block of code that performs a specific task. In Python, you can define a function using the def keyword, followed by the function name and parentheses () containing any parameters the function might take. The code block within the function is indented.

For example:

```python
def greet(name):
    print(f"Hello, {name}!")
```

Here, greet is a function that takes one parameter, name, and prints a greeting message.

Using Functions in Google Colab
To use functions in Google Colab, you can define them in a code cell and then call them from the same cell or another cell in the notebook.

For example, create a new code cell in your Google Colab notebook and type the following code:

```python
# Defining a function
def calculate_area(width, height):
    area = width * height
    return area

# Using the function
rectangle_area = calculate_area(10, 20)
print("Area of the rectangle:", rectangle_area)
```

After typing the code, press the "Run" button or use the keyboard shortcut Shift + Enter to execute the cell. You should see the output displayed below the cell: Area of the rectangle: 200

Parameters and Return Values
Functions can take parameters as input and return a value as output. In the example above, the calculate_area function takes two parameters, width and height, and returns the area of a rectangle.

You can also define functions with default parameter values:

```python
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Alice"))
print(greet("Bob", greeting="Hi"))
```
