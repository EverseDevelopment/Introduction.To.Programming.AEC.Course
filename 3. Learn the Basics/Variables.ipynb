Introduction to Variables in Python
In this tutorial, we'll learn about variables in Python and how to use them in Google Colab. Variables are essential in any programming language as they allow you to store and manipulate data.

What is a Variable?
A variable is a named location in memory that is used to store data. In Python, variables are created when you assign a value to them. For example:

```python
x = 5
name = "John"
print("Name:", name)
```

Here, x is a variable that stores the integer 5, and name is a variable that stores the string "John".

```python
# Creating variables
age = 25
height = 5.9
is_student = True
name = "Alice"

# Printing the variables
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is student:", is_student)
```

After typing the code, press the "Run" button or use the keyboard shortcut Shift + Enter to execute the cell. You should see the output displayed below the cell:

Name: Alice
Age: 25
Height: 5.9
Is student: True

Modifying Variables
Variables in Python can be modified after they are created. You can assign a new value to an existing variable, and the variable will now refer to the new value. Like the example below:

```python
# Modifying variables
age = 30
name = "Bob"

print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is student:", is_student)
```
