# Python Data Types

Python has several built-in data types that are commonly used to store and manipulate data. Here's a brief overview of the most commonly used data types in Python:


### Integer (int):
Used to represent whole numbers without a fractional part.

```python

### Integer (int):
integer_example = 10
print(f"<class 'int'> {integer_example}")
```

### Float (float):

Are used to represent real numbers, which are numbers that can have a fractional *part*

```python
### Float (float):
float_example = 10.5
print(f"<class 'float'> {float_example}")
```

### String (string):

Are used to represent text strings

```python
### String (str):
string_example = "Hello, Python!"
print(f"<class 'str'> {string_example}")
```

### Boolean (bool):

Booleans are a fundamental data type in programming used to represent truth values

```python
## 5. Boolean Type (bool):
bool_example = True
print(f"<class 'bool'> {bool_example}")
```
